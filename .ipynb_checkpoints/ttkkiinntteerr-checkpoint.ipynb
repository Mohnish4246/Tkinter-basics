# Tkinter
    Tkinter is Python’s standard GUI (Graphical User Interface) library.
    It is used to create desktop applications with windows, buttons, labels, text boxes, menus, etc.

    In simple words:
    Tkinter lets Python programs interact with users using windows instead of the terminal.
    
    Without Tkinter:
        User types input in terminal
        Output is printed as text
        
    With Tkinter:
        User clicks buttons
        Enters text in boxes
        Sees results in windows
### Uses of Tkinter
Tkinter is used to build:
* Calculator applications
* Form-based apps (Login, Registration)
* Small games
* ATM simulations
* Desktop tools for data analysis
* Student projects & mini projects
### Key Features
* Comes pre-installed with Python
* Easy to learn for beginners
* Event-driven (works on button clicks)
* Lightweight and fast
* Cross-platform (Windows, macOS, Linux)
### Common Tkinter components
| Component    | Use                      |
| ------------ | ------------------------ |
| `Tk()`       | Creates main window      |
| `Label`      | Displays text            |
| `Button`     | Performs action on click |
| `Entry`      | Takes user input         |
| `Text`       | Multi-line text          |
| `Frame`      | Groups widgets           |
| `messagebox` | Shows pop-up messages    |

```python
import tkinter as tk

root = tk.Tk()
root.title("My First App")
root.geometry("300x200")

label = tk.Label(root, text="Hello Students!", font=("Arial", 16))
label.pack()

entry = tk.Entry(root)
entry.pack()
# Button Click Event
# Introduce functions and event handling:
def show_text():
    value = entry.get() #Ankit
    label.config(text=f"Hello {value}")

btn = tk.Button(root, text="Submit", command=show_text)
btn.pack()

root.mainloop()
```

### pack(), grid(), and place()
    They are **layout managers** in Tkinter.
    Their job is to decide where widgets (buttons, labels, entries, etc.) appear inside a window.
    Without a layout manager, widgets will not show.
1. **pack() :**
    * Simple stacking (Automatic layout)
    * Places widgets one after another
    * Aligns them TOP / BOTTOM / LEFT / RIGHT
    * Automatically adjusts space
    * Example : Stacking books in a bag — one goes on top of another.
* Real-World Example : A welcome screen:
  * Logo on top
  * Text in middle
  * Button at bottom

```python
import tkinter as tk

root = tk.Tk()

tk.Label(root, text="Welcome").pack()
tk.Button(root, text="Login").pack()
tk.Button(root, text="Exit").pack()

root.mainloop()
```

2. **grid() :**
* Row & Column system (Table layout)
* What it does
  * Places widgets in rows and columns
  * Like an Excel sheet
  * Very organized and structured
* Think of it like:
  * An office form or exam answer sheet.
* Why we use it:
  * Perfect alignment
  * Best for forms
  * Clean professional UI
* Real-World Example:
  * Student Registration Form
  * Name:   [_____]
  * Age:    [_____]
  * Marks:  [_____]

```python
import tkinter as tk

root = tk.Tk()
root.geometry("200x100")

tk.Label(root, text="Name").grid(row=0, column=0,padx=5, pady=5)
tk.Entry(root).grid(row=0, column=1,padx=5, pady=5)

tk.Label(root, text="Age").grid(row=1, column=0,padx=5,pady=5)
tk.Entry(root).grid(row=1, column=1,padx=5,pady=5)

tk.Button(root, text="Submit").grid(row=2, column=1,padx=5,pady=5)

root.mainloop()
```

3. **place() :**
* Exact positioning (Absolute layout)
* What it does
  * Places widgets using x and y coordinates
  * Gives full control
* Think of it like:
  * Placing furniture in a room using a ruler
* Why we use it
  * Pixel-perfect positioning
  * Used in games or custom designs
* Real-World Example
  * A game score or floating button

```python
import tkinter as tk

root = tk.Tk()
root.geometry("300x200")

tk.Button(root, text="Click").place(x=50, y=80)

root.mainloop()
```

| Feature           | pack()     | grid()               | place()           |
| ----------------- | ---------- | -------------------- | ----------------- |
| Layout style      | Automatic  | Table (rows/columns) | Absolute          |
| Difficulty        | Very easy  | Medium               | Hard              |
| Best for          | Simple UI  | Forms                | Custom designs    |
| Responsive        | Yes        | Yes                  | No                |
| Used in real apps | Small apps | Most apps            | Games / custom UI |


```python
# Use Frame to separate pack() and grid()
import tkinter as tk

root = tk.Tk()

frame = tk.Frame(root)
frame.pack(pady=20)

tk.Label(frame, text="Username").grid(row=0, column=0, padx=10, pady=5)
tk.Entry(frame).grid(row=0, column=1, padx=10, pady=5)

tk.Label(frame, text="Password").grid(row=1, column=0, padx=10, pady=5)
tk.Entry(frame, show="*").grid(row=1, column=1, padx=10, pady=5)

root.mainloop()
```

```python
# simple calculator
import tkinter as tk

# ----------------- FUNCTIONS -------------------
def button_click(value):
    current = entry_box.get()
    entry_box.delete(0, tk.END)
    entry_box.insert(0, current + value)

def clear():
    entry_box.delete(0, tk.END)

def calculate():
    try:
        expression = entry_box.get()
        result = eval(expression)   # evaluates string expression
        entry_box.delete(0, tk.END)
        entry_box.insert(0, str(result))
    except:
        entry_box.delete(0, tk.END)
        entry_box.insert(0, "Error")
5
# ----------------- WINDOW -----------------------
root = tk.Tk()
root.title("Simple Calculator")
root.geometry("300x300")

# ----------------- ENTRY BOX --------------------
entry_box = tk.Entry(root, font=("Arial", 20), bd=5, justify="right")
entry_box.pack(fill="both", pady=10, padx=10)

# ------------- FRAME FOR BUTTONS ----------------
frame = tk.Frame(root)
frame.pack()

# All calculator buttons in a list (row-wise)
buttons = [
    "7", "8", "9", "/",
    "4", "5", "6", "*",
    "1", "2", "3", "-",
    "0", "C", "=", "+"
]

row = 0
col = 0

# ----------------- CREATE BUTTONS ---------------
for button in buttons:
    action = lambda x=button: button_click(x)

    if button == "C":
        tk.Button(frame, text=button, width=6, height=2, bg="red",
                  command=clear).grid(row=row, column=col, padx=5, pady=5)

    elif button == "=":
        tk.Button(frame, text=button, width=6, height=2, bg="lightgreen",
                  command=calculate).grid(row=row, column=col, padx=5, pady=5)

    else:
        tk.Button(frame, text=button, width=6, height=2,
                  command=action).grid(row=row, column=col, padx=5, pady=5)

    col += 1
    if col > 3:
        col = 0
        row += 1

# ----------------- START WINDOW ------------------
root.mainloop()
```

```python
import tkinter as tk

# ====== FUNCTIONS ======
def button_click(value):
    """
    This function runs whenever a number or operator button is clicked.
    It appends the clicked value to the Entry box.
    """
    current = entry.get()          # get current text in entry
    entry.delete(0, tk.END)        # clear entry
    entry.insert(0, current + value)  # insert new text (old + new char)

def clear_entry():
    """Clear the entry display."""
    entry.delete(0, tk.END)

def calculate_result():
    """
    Evaluate the expression in the entry.
    For now we use eval() for simplicity.
    (Later you can replace eval with a safer method.)
    """
    expression = entry.get()
    try:
        result = eval(expression)  # e.g. "2+3*4"
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except Exception:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

# ====== MAIN WINDOW ======
root = tk.Tk()
root.title("Simple Calculator")
root.geometry("300x400")

# ===== DISPLAY =====
entry = tk.Entry(root, font=("Arial", 20), borderwidth=5, justify="right")
entry.pack(padx=10, pady=10, fill="x")

# ===== FRAME FOR BUTTONS =====
button_frame = tk.Frame(root)
button_frame.pack()

# ===== BUTTONS WITH COMMANDS =====
# Row 0
btn_7 = tk.Button(button_frame, text="7", width=5, height=2,
                  command=lambda: button_click("7"))
btn_8 = tk.Button(button_frame, text="8", width=5, height=2,
                  command=lambda: button_click("8"))
btn_9 = tk.Button(button_frame, text="9", width=5, height=2,
                  command=lambda: button_click("9"))
btn_div = tk.Button(button_frame, text="/", width=5, height=2,
                    command=lambda: button_click("/"))

btn_7.grid(row=0, column=0, padx=5, pady=5)
btn_8.grid(row=0, column=1, padx=5, pady=5)
btn_9.grid(row=0, column=2, padx=5, pady=5)
btn_div.grid(row=0, column=3, padx=5, pady=5)

# Row 1
btn_4 = tk.Button(button_frame, text="4", width=5, height=2,
                  command=lambda: button_click("4"))
btn_5 = tk.Button(button_frame, text="5", width=5, height=2,
                  command=lambda: button_click("5"))
btn_6 = tk.Button(button_frame, text="6", width=5, height=2,
                  command=lambda: button_click("6"))
btn_mul = tk.Button(button_frame, text="*", width=5, height=2,
                    command=lambda: button_click("*"))

btn_4.grid(row=1, column=0, padx=5, pady=5)
btn_5.grid(row=1, column=1, padx=5, pady=5)
btn_6.grid(row=1, column=2, padx=5, pady=5)
btn_mul.grid(row=1, column=3, padx=5, pady=5)

# Row 2
btn_1 = tk.Button(button_frame, text="1", width=5, height=2,
                  command=lambda: button_click("1"))
btn_2 = tk.Button(button_frame, text="2", width=5, height=2,
                  command=lambda: button_click("2"))
btn_3 = tk.Button(button_frame, text="3", width=5, height=2,
                  command=lambda: button_click("3"))
btn_minus = tk.Button(button_frame, text="-", width=5, height=2,
                      command=lambda: button_click("-"))

btn_1.grid(row=2, column=0, padx=5, pady=5)
btn_2.grid(row=2, column=1, padx=5, pady=5)
btn_3.grid(row=2, column=2, padx=5, pady=5)
btn_minus.grid(row=2, column=3, padx=5, pady=5)

# Row 3
btn_clear = tk.Button(button_frame, text="C", width=5, height=2,
                      command=clear_entry)
btn_0 = tk.Button(button_frame, text="0", width=5, height=2,
                  command=lambda: button_click("0"))
btn_equal = tk.Button(button_frame, text="=", width=5, height=2,
                      command=calculate_result)
btn_plus = tk.Button(button_frame, text="+", width=5, height=2,
                     command=lambda: button_click("+"))

btn_clear.grid(row=3, column=0, padx=5, pady=5)
btn_0.grid(row=3, column=1, padx=5, pady=5)
btn_equal.grid(row=3, column=2, padx=5, pady=5)
btn_plus.grid(row=3, column=3, padx=5, pady=5)

root.mainloop()
```

#### Number Guessing Game

```python
import tkinter as tk
import random
attempts = 0
def start_new_game():
    global secret_number, attempts
    secret_number = random.randint(1, 100)

    msg_label.config(text="I chose a number between 1 and 100. Start guessing!")
    guess_entry.delete(0, tk.END)

def check_guess():
    global attempts
    guess_text = guess_entry.get()

    if not guess_text.isdigit():
        msg_label.config(text="Please enter a valid integer.")
        return

    guess = int(guess_text)
    attempts += 1

    if guess < secret_number:
        msg_label.config(text=f"Too low! Attempts: {attempts}")
    elif guess > secret_number:
        msg_label.config(text=f"Too high! Attempts: {attempts}")
    else:
        msg_label.config(text=f"🎉 Correct! Number was {secret_number}. Attempts: {attempts}")

# ==== MAIN WINDOW ====
root = tk.Tk()
root.title("Number Guessing Game")
root.geometry("350x200")

title_label = tk.Label(root, text="Number Guessing Game", font=("Arial", 14, "bold"))
title_label.pack(pady=5)

msg_label = tk.Label(root, text="Click 'New Game' to start.", wraplength=320)
msg_label.pack(pady=5)

frame = tk.Frame(root)
frame.pack(pady=5)

tk.Label(frame, text="Your Guess:").grid(row=0, column=0, padx=5)
guess_entry = tk.Entry(frame, width=10)
guess_entry.grid(row=0, column=1, padx=5)

check_btn = tk.Button(frame, text="Check", command=check_guess)
check_btn.grid(row=0, column=2, padx=5)

new_game_btn = tk.Button(root, text="New Game", command=start_new_game)
new_game_btn.pack(pady=5)

# start first game
start_new_game()

root.mainloop()
```

#### Simple ATM System

```python
import tkinter as tk
from tkinter import messagebox

balance = 1000.0  # starting balance

def show_balance():  # .config() is used to change the properties of a widget AFTER it has been created.
    result_label.config(text=f"Current Balance: ₹{balance:.2f}") # Update or modify an existing widget

def deposit_money():
    global balance
    amt_str = amount_entry.get()

    try:
        amount = float(amt_str)
        if amount <= 0:
            result_label.config(text="Enter a positive amount to deposit.")
            return
        balance += amount
        result_label.config(text=f"Deposited ₹{amount:.2f}. New Balance: ₹{balance:.2f}")
        amount_entry.delete(0, tk.END)
    except ValueError:
        result_label.config(text="Please enter a valid number.")

def withdraw_money():
    global balance
    amt_str = amount_entry.get()

    try:
        amount = float(amt_str)
        if amount <= 0:
            result_label.config(text="Enter a positive amount to withdraw.")
            return
        if amount > balance:
            result_label.config(text="Insufficient balance!")
            return
        balance -= amount
        result_label.config(text=f"Withdrew ₹{amount:.2f}. New Balance: ₹{balance:.2f}")
        amount_entry.delete(0, tk.END)
    except ValueError:
        result_label.config(text="Please enter a valid number.")

def exit_atm():
    if messagebox.askyesno("Exit", "Do you really want to exit?"):
        root.destroy()

# ==== MAIN WINDOW ====
root = tk.Tk()
root.title("Simple ATM System")
root.geometry("400x250")

title_label = tk.Label(root, text="Simple ATM System", font=("Arial", 16, "bold"))
title_label.pack(pady=5)

frame = tk.Frame(root)
frame.pack(pady=10)

tk.Label(frame, text="Amount (for deposit/withdraw):").grid(row=0, column=0, padx=5)
amount_entry = tk.Entry(frame, width=15)
amount_entry.grid(row=0, column=1, padx=5)

btn_frame = tk.Frame(root)
btn_frame.pack(pady=10)

tk.Button(btn_frame, text="Check Balance", width=15, command=show_balance).grid(row=0, column=0, padx=5, pady=5)
tk.Button(btn_frame, text="Deposit", width=15, command=deposit_money).grid(row=0, column=1, padx=5, pady=5)
tk.Button(btn_frame, text="Withdraw", width=15, command=withdraw_money).grid(row=1, column=0, padx=5, pady=5)
tk.Button(btn_frame, text="Exit", width=15, command=exit_atm).grid(row=1, column=1, padx=5, pady=5)

result_label = tk.Label(root, text="Welcome to the ATM.")
result_label.pack(pady=5)

root.mainloop()

# | Concept                 | Use                                    |
# | ----------------------- | -------------------------------------- |
# | `.config()`             | Update widget properties dynamically   |
# | `global balance`        | Modify global variable inside function |
# | `messagebox.askyesno()` | Popup Yes/No confirmation              |
# | `root.destroy()`        | Close entire Tkinter app               |
```

#### Prime number checker

```python
import tkinter as tk

def is_prime(num: int) -> bool:
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def check_prime():
    n_str = num_entry.get()

    if not n_str.isdigit():
        result_label.config(text="Please enter a valid positive integer.")
        return

    n = int(n_str)
    if is_prime(n):
        result_label.config(text=f"{n} is a PRIME number.")
    else:
        result_label.config(text=f"{n} is NOT a prime number.")

# ==== MAIN WINDOW ====
root = tk.Tk()
root.title("Prime Number Checker")
root.geometry("300x180")

title_label = tk.Label(root, text="Prime Number Checker", font=("Arial", 14, "bold"))
title_label.pack(pady=5)

frame = tk.Frame(root)
frame.pack(pady=10)

tk.Label(frame, text="Enter a number:").grid(row=0, column=0, padx=5)
num_entry = tk.Entry(frame, width=10)
num_entry.grid(row=0, column=1, padx=5)

check_btn = tk.Button(root, text="Check Prime", command=check_prime)
check_btn.pack(pady=5)

result_label = tk.Label(root, text="")
result_label.pack(pady=5)

root.mainloop()
```
